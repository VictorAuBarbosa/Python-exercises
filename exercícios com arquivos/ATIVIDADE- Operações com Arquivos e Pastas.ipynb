#### Exercício 3:
Escreva um programa que crie um novo diretório usando os.mkdir()

```python
from pathlib import Path
import os

arquivos = Path('C:\\Users\\Aluno 24\\Desktop\\Jupyter\\exercícios com arquivos')

if not os.path.exists(arquivos):
    os.mkdir(arquivos)
    print(f"Diretório '{arquivos}' foi criado com sucesso usando os.mkdir.")

subpasta = os.path.join(arquivos, 'Arquivos')

if not os.path.exists(subpasta):
    os.mkdir(subpasta)
    print(f"Diretório '{subpasta}' foi criado com sucesso dentro de '{arquivos}' usando os.mkdir.")
else:
    print(f"Diretório '{subpasta}' já existe dentro de '{arquivos}'.")
```

#### Exercício 5:
Crie um programa que leia o conteúdo de um arquivo e o imprima na tela usando open()

```python
import os

arquivos = 'C:\\Users\\Aluno 24\\Desktop\\Jupyter\\exercícios com arquivos\\Arquivos'

if os.path.exists(arquivos):
    conteudo = os.listdir(arquivos)

    for item in conteudo:
        if os.path.isfile(os.path.join(arquivos, item)):
            print(f"Arquivo encontrado: {item}")
            
            with open(os.path.join(arquivos, item), 'r') as arquivo:
                conteudo_arquivo = arquivo.read()
                print(f"Conteúdo do arquivo {item}:\n{conteudo_arquivo}")
else:
    print('Não tem nada ou não extiste esse diretório.')
```

#### Exercício 6:
Crie um programa que escreva uma lista de strings em um arquivo de texto usando open()

```python
lista_strings = [
    "Não sei quem sou",
    "Eu sou alguém?",
    "Eu mereço ser alguém..."
]

caminho_arquivo = 'C:\\Users\\Aluno 24\\Desktop\\Jupyter\\exercícios com arquivos\\Arquivos\\Coisas.txt'

with open(caminho_arquivo, 'w') as arquivo:
    for linha in lista_strings:
        arquivo.write(linha + '\n')

print("Lista de strings foi escrita no arquivo com sucesso.")
```

#### Exercício 11:
Crie um programa que liste todos os arquivos em um diretório específico usando a biblioteca pathlib com o método iterdir()

```python
import pathlib

def listar_arquivos_em_diretorio(caminho_diretorio):
    caminho_diretorio = pathlib.Path(caminho_diretorio)
    
    if caminho_diretorio.is_dir():
        for item in caminho_diretorio.iterdir():
            if item.is_file():
                print(f"Arquivo: {item}")
    else:
        print(f"'{caminho_diretorio}' não é um diretório válido.")

caminho_diretorio = 'C:\\Users\\Aluno 24\\Desktop\\Jupyter\\teste'
listar_arquivos_em_diretorio(caminho_diretorio)
```
