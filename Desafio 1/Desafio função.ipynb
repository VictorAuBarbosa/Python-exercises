## Tarefas:

## 1 - Registro de Empréstimos:

### Registre os empréstimos de livros para os  alunos fictícios da escola.
## 2 -  Avaliação de Livros:

### Após os empréstimos, simule que os alunos leram os livros e atribua notas de avaliação aos livros emprestados.

## 3 - Avaliação do Atendimento da Biblioteca:

###  Registre avaliações do atendimento da biblioteca por parte dos alunos.

## 4 - Imprimir Relatório:

### Após realizar as ações anteriores, imprima um relatório que apresente as informações sobre os empréstimos, avaliações dos livros e a média das avaliações do atendimento da biblioteca.


```python
emprestimos = []
avaliacoes_livros = []
avaliacoes_atendimento = []

def registrar_emprestimo():
    aluno = input("Nome do aluno: ")
    livro = input("Título do livro emprestado: ")
    emprestimo = {"aluno": aluno, "livro": livro}
    emprestimos.append(emprestimo)
```

```python
def atribuir_avaliacao_livro():
    livro = input("Digite o nome do livro a ser avaliado: ")
    nota = float(input("Digite a nota de avaliação do livro (0 a 5): "))
    avaliacao = {'livro': livro, 'nota': nota}
    avaliacoes_livros.append(avaliacao)
```

```python
def registrar_avaliacao_atendimento():
    atendimento = input("Digite uma avaliação do atendimento da biblioteca: ")
    nota = float(input("Digite a nota de avaliação do atendimento (0 a 5): "))
    avaliacao = {'atendimento': atendimento, 'nota': nota}
    avaliacoes_atendimento.append(avaliacao)

def calcular_media_avaliacoes_atendimento():
    if len(avaliacoes_atendimento) == 0:
        return 0
    soma_notas = sum(avaliacao['nota'] for avaliacao in avaliacoes_atendimento)
    media = soma_notas / len(avaliacoes_atendimento)
    return media
```

```python
def imprimir_relatorio():
    print("### Empréstimos de Livros ###")
    for emprestimo in emprestimos:
        print(f"Aluno: {emprestimo['aluno']}, Livro: {emprestimo['livro']}")

    print("\n### Avaliações dos Livros ###")
    for avaliacao in avaliacoes_livros:
        print(f"Livro: {avaliacao['livro']}, Nota: {avaliacao['nota']}")

    print(f"\n### Média das Avaliações do Atendimento da Biblioteca: {calcular_media_avaliacoes_atendimento()} ###")

while True:
    print("\nEscolha uma opção:")
    print("1 - Registrar Empréstimo de Livro")
    print("2 - Atribuir Avaliação a Livro")
    print("3 - Registrar Avaliação do Atendimento da Biblioteca")
    print("4 - Imprimir Relatório")
    print("5 - Sair")

    escolha = input("Digite o número da opção desejada: ")

    if escolha == "1":
        registrar_emprestimo()
    elif escolha == "2":
        atribuir_avaliacao_livro()
    elif escolha == "3":
        registrar_avaliacao_atendimento()
    elif escolha == "4":
        imprimir_relatorio()
    elif escolha == "5":
        break
    else:
        print("Opção inválida. Por favor, escolha uma opção válida.")
```
